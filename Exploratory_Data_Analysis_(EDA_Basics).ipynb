# Assignment Questions

# Q1. What is the average, median, and standard deviation of house prices?

Ans:

```python
avg_price = _df['price'].mean()
median_price = _df['price'].median()
std_price = _df['price'].std()

print("Average Price:", avg_price)
print("Median Price:", median_price)
print("Standard Deviation:", std_price)
```

### Interpretation

* Average price shows the overall average cost of houses.
* Median represents the middle value and helps understand skewness.
* Standard deviation indicates how spread out the prices are.
* If standard deviation is high, house prices vary significantly.



# Q2. Which number of bedrooms is most common?

Ans:

```python
most_common_bedrooms = _df['bedrooms'].mode()[0]
print("Most Common Bedrooms:", most_common_bedrooms)
```

### Interpretation

* The mode tells us the most frequently occurring number of bedrooms.
* This indicates the type of houses most preferred in the market.



# Q3. Check if the price distribution is skewed.

Ans:

```python
skewness = _df['price'].skew()
print("Skewness:", skewness)

sns.histplot(_df['price'], kde=True)
plt.title('Price Distribution')
plt.show()
```

### Interpretation

* If skewness > 0 → Positively skewed.
* If skewness < 0 → Negatively skewed.
* Housing prices are usually positively skewed because a few luxury homes have extremely high prices.



# Q4. What is the average price for each number of bedrooms?

Ans:

```python
avg_price_bedrooms = _df.groupby('bedrooms')['price'].mean()
print(avg_price_bedrooms)
```

### Interpretation

* This shows how house price changes with bedroom count.
* Usually, houses with more bedrooms have higher prices.



# Q5. What is the relationship between living area and price?

(Use: area without Basement)

Ans:

```python
correlation = _df['living_area'].corr(_df['price'])
print("Correlation:", correlation)

sns.scatterplot(x=_df['living_area'], y=_df['price'])
plt.title('Living Area vs Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

### Interpretation

* Positive correlation means larger houses tend to have higher prices.
* If correlation is close to 1, relationship is strong.
* Scatterplot helps visualize the trend.



# Q6. Identify any anomalies where houses have high prices but low areas.

Ans:
```python
anomalies = _df[(_df['price'] > _df['price'].quantile(0.90)) &
                (_df['living_area'] < _df['living_area'].quantile(0.25))]

print(anomalies[['price', 'living_area']])
```

### Interpretation

* These are houses with unusually high prices despite smaller areas.
* Possible reasons:

  * Prime location
  * Luxury interiors
  * Waterfront or premium facilities



# Q7. Compare average prices based on number of floors and houses with or without waterfront.

Ans:

```python
comparison = _df.groupby(['floors', 'waterfront'])['price'].mean()
print(comparison)
```

### Interpretation

* Houses with waterfront generally have higher prices.
* Multi-floor houses may also cost more due to larger construction size.



# Q8. Identify the minimum and maximum house price. What does this indicate?

Ans:
```python
min_price = _df['price'].min()
max_price = _df['price'].max()

print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
```

### Interpretation

* Minimum price represents the most affordable house.
* Maximum price represents luxury or premium property.
* A large gap between minimum and maximum suggests a diverse housing market.



# Q9. Which location (zipcode/area) has the highest average price?

Ans:

```python
highest_location = _df.groupby('zipcode')['price'].mean().sort_values(ascending=False)

print(highest_location.head(1))
```

### Interpretation

* The zipcode with the highest average price is considered the premium location.
* These areas may have better facilities, infrastructure, or demand.
